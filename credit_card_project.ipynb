```python
1] Transaction_Analysis
```

```python
total_transactions = len(df)
print("Total Transactions:", total_transactions)
```

```python
fraud_transactions = df[df["Class"] == 1].shape[0]
print("Fraud Transactions:", fraud_transactions)
```

```python
fraud_percentage = (fraud_transactions / total_transactions) * 100
print("Fraud Percentage:", round(fraud_percentage, 2), "%")
```

```python
print("Median Transaction:", df["Amount"].median())
```

```python
3. Time Analysis
```

```python
df["Hour"] = (df["Time"] // 3600) % 24
```

```python
fraud_hour = df[df["Class"] == 1].groupby("Hour").size()
print(fraud_hour)
```

```python


fraud_hour.plot(kind="bar")
plt.title("Fraud by Hour")
plt.xlabel("Hour")
plt.ylabel("Fraud Count")
plt.show()
```

```python
peak_hour = fraud_hour.idxmax()
print("Peak Fraud Hour:", peak_hour)
```
